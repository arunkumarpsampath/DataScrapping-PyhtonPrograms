```python
from selenium import webdriver
```

```python
browser = webdriver.Chrome('chromedriver.exe')
```

```python
browser.get('http://www.elections.in/')
```

```python
import pandas as pd,numpy as np
```

```python
constituencywise = pd.DataFrame()
```

```python
prev=0
present=1
while(1):
    text = browser.find_element_by_css_selector('div.mapcontent').text
    try:
        present = text.split('Constituency : ')[1].split('\n')[0]
        
        if prev != present:

                cur_row = {'constituency': present,
                           'State' : text.split('Constituency : ')[1].split('\nState : ')[1].split('\n')[0],
                           'Candidate' : text.split('Constituency : ')[1].split('\nState : ')[1].split('\n')[1].split('\n')[0],
                           'Party' : text.split('\n')[-1]
                          }

                constituencywise = constituencywise.append(cur_row,ignore_index=True)
                prev = text.split('Constituency : ')[1].split('\n')[0]
    except Exception as e:
        print(e)
```
