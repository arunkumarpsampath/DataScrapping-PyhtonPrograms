```python
import pandas as pd,numpy as np
from selenium import webdriver
browser = webdriver.Chrome('chromedriver.exe')
```

```python
browser.get('https://www.goodreads.com/choiceawards/best-books-2017')
```

```python
categoriestag = browser.find_element_by_css_selector('div#categories.clearFix')
```

```python
categorylisttag = categoriestag.find_elements_by_css_selector('div.category.clearFix')
bestbooks2017 = pd.DataFrame()
for tag in categorylisttag:
    cur_row = {
        "category": tag.find_element_by_css_selector('h4.category__copy').text,
        "Book name": tag.find_element_by_css_selector('img.category__winnerImage').get_attribute('alt'),
        "Cover URL": tag.find_element_by_css_selector('img.category__winnerImage').get_attribute('src'),
        "Description URL": tag.find_element_by_css_selector('a').get_attribute('href')
    }
    bestbooks2017 = bestbooks2017.append(cur_row,ignore_index=True)
bestbooks2017
```

```python
import time,re
bookinfo = pd.DataFrame()
for index,data in bestbooks2017.iterrows():
    browser.get(data['Description URL'])
    time.sleep(2)
    parent_tag = browser.find_element_by_css_selector('div.winner.stacked')
    cur_row = {
           "authorname" : parent_tag.find_element_by_css_selector('div.authorName__container > a.authorName').text,
           "description" : parent_tag.find_element_by_css_selector('div#description').text,
           "Votes obtained" : int(re.findall("^([^ ]*) ",parent_tag.find_element_by_css_selector('div.gcaWinnerHeader > span.gcaNumVotes').text)[0].replace(',',"")),
           "Total Votes" : int(re.findall("^([^ ]*) ",browser.find_element_by_css_selector('h2.gcaSubheader > div.greyText').text)[0].replace(',',""))
    }
    bookinfo = bookinfo.append(cur_row,ignore_index = True)
bookinfo
```

```python
bookinfo['% votes for the winner'] = bookinfo["Votes obtained"] * 100 / bookinfo["Total Votes"]
bookinfo
```

```python
bestbooks2017.join(bookinfo)
```
