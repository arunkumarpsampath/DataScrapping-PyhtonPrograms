```python
from selenium import webdriver
browser = webdriver.Chrome('chromedriver.exe')
```

```python
import pandas as pd
```

```python
#Naukri website scrapping
browser.get("https://www.naukri.com/data-science-jobs")
```

```python
parent_tag = browser.find_elements_by_css_selector('div[type="tuple"]')
#len(parent_tag)
df_paras = pd.DataFrame()
for tag in parent_tag:
    cur_row = {'Job URL':tag.get_attribute('data-url'),
               'Job id':tag.get_attribute('id'),
               'Title':tag.find_element_by_css_selector('li.desig').text,
               'Company':tag.find_element_by_css_selector('span.org').text,
               'Experience':tag.find_element_by_css_selector('span.exp').text,
               'Location':tag.find_element_by_css_selector('span.loc').text,
               'Key skills':tag.find_element_by_css_selector('span.skill').text,
               'Salary':tag.find_element_by_css_selector('span.salary').text,
               'Date':tag.find_element_by_css_selector('span.date').text
              }
    df_paras = df_paras.append(cur_row, ignore_index=True)
df_paras
```

```python
import time
def find(tag,searchtag):
    try: return tag.find_element_by_css_selector(searchtag).text
    
    except: return None  
df_paras = pd.DataFrame()
for i in range(1,10):
    url = 'https://www.naukri.com/data-science-jobs-%d' % (i)
    browser.get(url)
    time.sleep(2)
    print(i)
    parent_tag = browser.find_elements_by_css_selector('div[type="tuple"]') 
    
    for tag in parent_tag:
        cur_row = {'Job URL':tag.get_attribute('data-url'),
                   'Job id':tag.get_attribute('id'),
                   'Title':find(tag,'li.desig'),
                   'Company':find(tag,'span.org'),
                   'Experience':find(tag,'span.exp'),
                   'Location':find(tag,'span.loc'),
                   'Key skills':find(tag,'span.skill'),
                   'Salary':find(tag,'span.salary'),
                   'Date':find(tag,'span.date')
                  }
        df_paras = df_paras.append(cur_row, ignore_index=True)
df_paras.to_csv('all_jobs_datascience.csv',index=False)
print('done')
```

```python
import time
def find(tag,searchtag):
    try: return tag.find_element_by_css_selector(searchtag).text
    
    except: return None  
url = 'https://www.naukri.com/data-science-jobs'
browser.get(url)
df_paras = pd.DataFrame()
while True:
    time.sleep(2)
    parent_tag = browser.find_elements_by_css_selector('div[type="tuple"]') 
    
    for tag in parent_tag:
        cur_row = {'Job URL':tag.get_attribute('data-url'),
                   'Job id':tag.get_attribute('id'),
                   'Title':find(tag,'li.desig'),
                   'Company':find(tag,'span.org'),
                   'Experience':find(tag,'span.exp'),
                   'Location':find(tag,'span.loc'),
                   'Key skills':find(tag,'span.skill'),
                   'Salary':find(tag,'span.salary'),
                   'Date':find(tag,'span.date')
                  }
        df_paras = df_paras.append(cur_row, ignore_index=True)
    try: 
        button = browser.find_elements_by_css_selector('button.grayBtn')[-1]
        if button.text == 'Next':
            button.click()
            time.sleep(2)
        else:
            break
    except Exception as e:
        break
df_paras.to_csv('all1_jobs_datascience.csv',index=False)
print('done')
```

```python
#OYO rooms
import time,re
def find(parent,tag,search='text'):
    try:
        if search == 'text':
            return parent.find_element_by_css_selector(tag).text
        elif search == 'price':
            try:
                return re.findall("(^. .{3}) .*",parent.find_element_by_css_selector(tag).text)[0]
            except:
                return re.findall("(^. .{4}) .*",parent.find_element_by_css_selector(tag).text)[0]
        elif search == 'aprice':
            try: 
                return re.findall("^. .{3} (. .{4}) .*",parent.find_element_by_css_selector(tag).text)[0]
            except:
                return re.findall("^. .{4} (. .{4}) .*",parent.find_element_by_css_selector(tag).text)[0]
        elif search == 'rating':
            return parent.find_element_by_css_selector(tag).text.split(' ')[0]
        else:
            return parent.find_element_by_css_selector(tag).get_attribute(search)
    except:
        return None
df = pd.DataFrame()
browser.get("https://www.oyorooms.com/oyos-in-bangalore?adults=1&checkin=06%2F03%2F2019&checkout=07%2F03%2F2019&children=0&city=bangalore&country=India&employee_id=null&guests=1&latitude=null&location=Bangalore%2CIndia&longitude=null&page=1&rooms=1&searchType=city&src=null")
while True:
    time.sleep(2)
    hotelstag = browser.find_elements_by_css_selector('div.newHotelCard')
    for hotel in hotelstag:
        cur_row = {'HotelName': find(hotel,'span.newHotelCard__hotelName'),
                'Hotel Address':find(hotel,'span.newHotelCard__hotelAddress'),
                'Actual Price':find(hotel,'div.newHotelCard__pricing','aprice'),
                'Discounted Price':find(hotel,'div.newHotelCard__pricing','price'),
                'Rating':find(hotel,'span.hotelRating__value'),
                'RatingCount':find(hotel,'div.hotelRating__count','rating'),
                'ImageUrl':find(hotel,'img.carousal__image','src')
              }
        
        df = df.append(cur_row,ignore_index=True)
    
    #To press next button
    try:
        atag = browser.find_elements_by_css_selector('a.disabled')[0]
        atag.get_attribute('div.btn-next')
        atag.click()
        print(1)
    except Exception as e:
        print(e)
        break
df.to_csv('bangalorehotels.csv', index=False)
df
```
