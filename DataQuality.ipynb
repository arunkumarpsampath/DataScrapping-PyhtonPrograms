```python
def exp_range(exp):
    if exp == 0:
        return 'Fresher'
    elif exp > 0 and exp <=2:
        return '0-2'
    elif exp > 2 and exp <=5:
        return '2-5'
    elif exp > 5 and exp <=10:
        return '5-10'
    elif exp > 10 and exp <=20:
        return '10-20'
    else:
        return 'above 20'

salaries['exp_range'] = salaries['min Experience'].apply(exp_range)
salaries.head()
```

```python
salaries['exp_range'].value_counts()
```

```python
(salaries.groupby('exp_range')['sal_min'].mean() /100000).sort_values()
```

# using digitise for interval creation

```python
a = [10,10,30,50,70,50]
bins = [30,60,80]
np.digitize(a,bins)
```

```python
#creating bins for salaries
bins = [0.1,2.1,5.1,10.1,20.1]
salaries['exp_bin_range'] = salaries['min Experience'].apply(lambda x: np.digitize(x,bins)).replace({
    0: 'Freshers',
    1: '0-2',
    2: '2-5',
    3: '5-10',
    4: '10-20',
    5: 'Above 20'
})
salaries['exp_bin_range'].head()
```

```python
salaries['exp_bin_range'].value_counts()
```

```python
(salaries.groupby('exp_bin_range')['sal_min'].mean() /100000).sort_values()
```

```python
salaries.groupby('exp_bin_range')['sal_min','sal_max'].mean()/100000
```
